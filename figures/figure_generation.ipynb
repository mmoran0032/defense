```python
x = np.linspace(0.1, 1)

m = 1
v = 1
q = 1

def Erho(m, v, q):
    return m * v**2 / q

def Brho(m, v, q):
    return m * v * q

def T_q(m, v, q):
    return m * v**2 / (2 * q)

# magnetic: f_b = f_c
def magnetic(T_q, Bp):
    return 0.5 * Bp / T_q

def electric(T_q, m_q):
    pass  # vertical line = Ep / 2

def wien(T_q, v):
    return (2 / v**2) * T_q

fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(x, magnetic(x, Brho(m, v, q)), c='C0', label='Magnetic')
ax.axvline(Erho(m, v, q) / 2, c='C1', label='Electric')
ax.plot(x, wien(x, v), c='C3', label='Wien')

sns.despine(ax=ax, left=True, bottom=True)
ax.get_xaxis().set_ticks([])
ax.get_yaxis().set_ticks([])
ax.set_xlabel('T/q')
ax.set_ylabel('m/q')
# ax.set_ylim((0, 1))
# ax.set_xlim((0, 1))
ax.legend(frameon=False, fontsize='xx-small')
fig.savefig('particle_selection_by_field.png', transparent=True)
plt.show()
```
